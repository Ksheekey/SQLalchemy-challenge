### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months
jun_dec = df
jun_dec['year'] = jun_dec['date'].dt.year
jun_dec['month'] = jun_dec['date'].dt.month
jun_dec['day'] = jun_dec['date'].dt.day
jun_dec_data = jun_dec.loc[(jun_dec["month"] == 6) | (
    jun_dec["month"] == 12), :]
#jun_data = jun_dec.loc[jun_dec.month == 6]
#jun_data
jun_dec_data
```

```python
# Identify the average temperature for June
jun_temp = jun_dec_data[['month','tobs']]
jun_temp_data = jun_temp.loc[(jun_temp["month"] == 6)]
june_temp_data = jun_temp_data[['tobs']]
June_avg_temp = round(june_temp_data['tobs'].mean(),2)
June_avg_temp
```

```python
# Identify the average temperature for December
dec_temp = jun_dec_data[['month','tobs']]
dec_temp_data = dec_temp.loc[(dec_temp["month"] == 12)]
decm_temp_data = dec_temp_data[['tobs']]
Decm_avg_temp = round(decm_temp_data['tobs'].mean(),2)
Decm_avg_temp
```
